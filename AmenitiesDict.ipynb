```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.dates as mdates
import matplotlib
```

```python
listings=pd.read_csv("combined2019fini.csv")
```

```python
conditionss = [
    (listings['busy'] <= .413999999),
    (listings['busy'] >= .414) & (listings['busy'] < .8499999),
    (listings['busy'] >= .85),
    (listings['busy']==0),
    ]
values=['Low','Mid','High','NA']
listings['busy_brackets']=np.select(conditionss,values)
```

```python
conditionss = [
    (listings['neighbourhood_cleansed'] == 'Long Beach')|(listings['neighbourhood_cleansed'] == 'Santa Monica')|(listings['neighbourhood_cleansed'] == 'Mar Vista')|(listings['neighbourhood_cleansed'] == 'Venice'),
    (listings['neighbourhood_cleansed'] == 'Downtown')|(listings['neighbourhood_cleansed'] == 'Hollywood')|(listings['neighbourhood_cleansed'] == 'Silver Lake')|(listings['neighbourhood_cleansed'] == 'Echo Park'),
    (listings['neighbourhood_cleansed'] == 'Pasadena')|(listings['neighbourhood_cleansed'] == 'Glendale')|(listings['neighbourhood_cleansed'] == 'Sherman Oaks')|(listings['neighbourhood_cleansed'] == 'Studio City'),
    ]
values=['Coastal','Urban','Suburban']
listings['geographical']=np.select(conditionss,values)
```

```python
reduced=listings[listings['geographical'].isin(values)]
```

```python
reduced.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1)
```

```python
for index, i in enumerate(reduced['amenitylist']):
    temp_cat={}
    for j in i:
        for k,v in dic.items():
            for amen in v:
                if j.lower() == amen.lower():
                    if k in temp_cat:
                        temp_cat[k] += 1
                    else:
                        temp_cat[k]=1
    for c, total in temp_cat.items():
        if c=='Bathroom':
            reduced['Bathroom_count'].iloc[index]=total
        if c=='Bedroom':
            reduced['Bedroom_count'].iloc[index]=total
        if c=='Childcare':
            reduced['Childcare_count'].iloc[index]=total
        if c=='Convenience':
            reduced['Convenience_count'].iloc[index]=total
        if c=='Entertainment':
            reduced['Entertainment_count'].iloc[index]=total
        if c=='Kitchen':
            reduced['Kitchen_count'].iloc[index]=total
        if c=='Luxury':
            reduced['Luxury_count'].iloc[index]=total
        if c=='Mobility':
            reduced['Mobility_count'].iloc[index]=total
        if c=='Outdoor':
            reduced['Outdoor_count'].iloc[index]=total
        if c=='Utility':
            reduced['Utility_count'].iloc[index]=total
        if c=='Waterfeature':
            reduced['Waterfeature_count'].iloc[index]=total
        if c=='Work':
            reduced['Work_count'].iloc[index]=total
```
