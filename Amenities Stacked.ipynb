```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.dates as mdates
import matplotlib
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
listings=pd.read_csv("combined2019fini.csv")
```

```python
conditionss = [
    (listings['busy'] <= .413),
    (listings['busy'] >= .414) & (listings['busy'] < .84),
    (listings['busy'] >= .85),
    (listings['busy']==0),
    ]
values=['Low','Mid','High','NA']
listings['busy_brackets']=np.select(conditionss,values)
```

```python
conditionss = [
    (listings['neighbourhood_cleansed'] == 'Long Beach')|(listings['neighbourhood_cleansed'] == 'Santa Monica')|(listings['neighbourhood_cleansed'] == 'Mar Vista')|(listings['neighbourhood_cleansed']=='Venice'),
    (listings['neighbourhood_cleansed'] == 'Downtown')|(listings['neighbourhood_cleansed'] == 'Hollywood')|(listings['neighbourhood_cleansed'] == 'Silver Lake')|(listings['neighbourhood_cleansed'] == 'Echo Park'),
    (listings['neighbourhood_cleansed'] == 'Pasadena')|(listings['neighbourhood_cleansed'] == 'Glendale')|(listings['neighbourhood_cleansed'] == 'Long Beach')|(listings['neighbourhood_cleansed'] == 'Sherman Oaks')|(listings['neighbourhood_cleansed'] == 'Studio City'),
    ]
values=['Coastal','Urban','Suburban']
listings['geographical']=np.select(conditionss,values)
```

```python
listings[listings['neighbourhood_cleansed']=='Venice']
```

```python
reduced=listings[listings['geographical'].isin(values)]
```

```python
reduced[reduced['geographical']=='Coastal']
```
